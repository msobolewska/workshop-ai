```python
import os
import pandas as pd
```

```python
folder_path = "stocks"

empty_files = []

for file in os.listdir(folder_path):
    file_path = os.path.join(folder_path, file)
    if os.path.isfile(file_path) and os.path.getsize(file_path) == 0:
        empty_files.append(file)

if empty_files:
    print("Puste pliki w folderze 'stocks':")
    for empty_file in empty_files:
        print(empty_file)
else:
    print("Brak pustych plików w folderze 'stocks'.")
```

```python
folder_path = "stocks"

files = os.listdir(folder_path)

dfs = []

for file in files:
    if file.endswith(".txt"):  
        file_path = os.path.join(folder_path, file)
        if os.path.getsize(file_path) > 0:
            df = pd.read_csv(file_path)
            dfs.append(df)

if dfs:
    merged_df = pd.concat(dfs, ignore_index=True)

    merged_file_path = "merged.csv"
    merged_df.to_csv(merged_file_path, index=False)

    print("Zmergowane dane zostały zapisane do pliku:", merged_file_path)
else:
    print("Brak danych do zmergowania lub wszystkie pliki są puste.")
```

```python
df = pd.read_csv("merged.csv")
df.head()
```

```python
df = df.drop(columns=['<PER>', '<TIME>', '<OPENINT>'])
```

```python
df = df.rename(columns={'<TICKER>': 'TICKER',
                        '<DATE>': 'DATE',
                        '<OPEN>': 'OPEN',
                        '<HIGH>': 'HIGH',  
                        '<LOW>': 'LOW',  
                        '<CLOSE>': 'CLOSE',
                        '<VOL>': 'VOL'})
df.head()
```

```python
df['DATE'] = pd.to_datetime(df['DATE'], format='%Y%m%d')
df['DATE'] = df['DATE'].dt.strftime('%m.%d.%Y')
df.head()
```

```python
print("Ilość tickerów:", df['TICKER'].nunique())
df.info()
```

```python
delta = (pd.to_datetime(df['DATE']).max() - pd.to_datetime(df['DATE']).min())
print("Przedział czasowy danych historycznych:\n", delta)
```

```python
mean_dict = {}

for key in df['TICKER'].unique():
    value = df[df['TICKER'] == key ]['VOL'].mean()
    mean_dict[key]= value


avaerage_s = pd.Series(mean_dict).transpose()
top10_s = avaerage_s.sort_values(ascending=False)[:10]

print("Top 10 spółek z najwyższą średnią wolumenu:", top10_s.index)
```

```python
print(df[df['TICKER'] == 'SAN'].tail())


# Wybór danych tylko dla pierwszych 10 spółek z najwyższą średnią wolumenu
top10_tickers = top10_s.index
top10_data = df[df['TICKER'].isin(top10_tickers)]

# Obliczenie minimalnej i maksymalnej daty dla wybranych spółek
min_date = top10_data['DATE'].min()
max_date = top10_data['DATE'].max()

print("Minimalna data notowań dla wybranych spółek:", min_date)
print("Maksymalna data notowań dla wybranych spółek:", max_date)
```

```python
# Dla tych 10 spółek dane uzupełniamy o wyniki finanasowe
ticker_to_name_dict = {'SAN': 'SANTANDER', 'GTN': 'GETIN', 'TPE': 'TAURONPE', 'OPL': 'ORANGEPL', 'PZU':'PZU', 'PKO': 'PKOBP',
                      'ALE': 'ALLEGRO', 'PGE': 'PGE', 'PKN': 'PKNORLEN', 'CEZ': 'CEZ'}

folder_path = "data_from_scrapper"

files = os.listdir(folder_path)

dfs_fin = []

for filename_short in top10_s.index:
    filename = filename_short
    if filename in ticker_to_name_dict:
        filename = ticker_to_name_dict[filename]
    else:
        print("Uwaga: Klucz '{}' nie znaleziony w słowniku.".format(filename))
    
    filename += ".csv"
    
    filepath = os.path.join(folder_path, filename)
    df_fin = pd.read_csv(filepath)

    # Ręcznie dodaj kolumnę z nazwą spółki
    df_fin['TICKER'] = filename_short
    dfs_fin.append(df_fin)

combined_df = pd.concat(dfs_fin, ignore_index=True)


if dfs_fin:
    merged_df = pd.concat(dfs_fin, ignore_index=True)
    merged_file_path = "merged_financial.csv"
    merged_df.to_csv(merged_file_path, index=False)
```

```python
df_fin = pd.read_csv("merged_financial.csv")
df_fin.head()
```

```python
# Zamiana formatu timestamp do współnego, by można było złączyć
df_fin['year'] = df_fin['kwartał'].str.split(' ').str[-1]
df_fin['quarter'] = df_fin['kwartał'].str.split(' ').str[0]
quarter_to_month = {'I': '01', 'II': '04', 'III': '07', 'IV': '10'}
df_fin['month'] = df_fin['quarter'].map(quarter_to_month)
df_fin['DATE'] = pd.to_datetime(df_fin['year'] + '-' + df_fin['month'] + '-01')
df_fin['DATE'] = df_fin['DATE'].dt.strftime('%m.%d.%Y')
df_fin.head()
```

```python
merged_df = pd.merge(df, df_fin, on=['DATE', 'TICKER'], how='outer')

merged_df.head()
```

```python
import matplotlib.pyplot as plt
from matplotlib.dates import YearLocator, DateFormatter

# Wybór danych tylko dla pierwszych 10 spółek z najwyższą średnią wolumenu
top10_tickers = top10_s.index
top10_data = df[df['TICKER'].isin(top10_tickers)]

# Iteracja przez każdą ze spółek z top 10 i rysowanie wykresów dla każdej spółki
for ticker in top10_tickers:
    ticker_data = top10_data[top10_data['TICKER'] == ticker]
    plt.figure()  # Tworzenie nowego wykresu
    plt.plot(ticker_data['DATE'], ticker_data['CLOSE'])
    plt.title(f'Cena zamknięcia dla spółki {ticker}')
    plt.xlabel('Data')
    plt.ylabel('Cena zamknięcia')

    # Formatowanie dat na osi x - tylko jedna data na rok
    ax = plt.gca()
    ax.xaxis.set_major_locator(YearLocator())
    ax.xaxis.set_major_formatter(DateFormatter('%Y'))
    plt.xticks(rotation=45)  # Obrót etykiet osi x dla lepszej czytelności dat

plt.tight_layout()  # Poprawa rozmieszczenia wykresów
plt.show()
```
