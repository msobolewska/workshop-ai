```python
import requests
from bs4 import BeautifulSoup
import csv
import re
import os
```

```python
# step 1
# get links to all companies
# links are of format: /inwestowanie/profile/quote.html?symbol=<company_name>
base_url = "https://www.bankier.pl/gielda/notowania/akcje"

page = requests.get(base_url)
page_content = page.content
soup = BeautifulSoup(page_content, 'html.parser')

hrefs = [a['href'] for a in soup.find_all('a') if 'inwestowanie' in a.get('href')]
print(hrefs)
```

```python
#step 2
# for each company navigate to its link, then go to "Wyniki finansowe" tab
# perform scrapping
# save result to a csv file

def remove_spaces(text):
    return ''.join(text.split())

for href in hrefs:
    full_url = "https://www.bankier.pl" + href
    page = requests.get(full_url)
    page_content = page.content
    soup = BeautifulSoup(page_content, 'html.parser')

    wyniki_href = [a['href'] for a in soup.find_all('a') if 'Wyniki finansowe' in a.get_text()][0]
    print(wyniki_href)

    # scrap data from "Wyniki finansowe" page table

    wyniki_full_url = "https://www.bankier.pl" + wyniki_href + "/skonsolidowany/kwartalny/standardowy/"

    page_number=1

    wyniki_page = requests.get(wyniki_full_url)
    wyniki_page_content = wyniki_page.content

    table_data = []
    table_data.append(column_names)

    while True:
        print(page_number)
        
        wyniki_fin_url = wyniki_full_url + str(page_number)
        page = requests.get(wyniki_fin_url)
        page_content = page.content
        soup = BeautifulSoup(page_content, 'html.parser')
        table = soup.table

        if table is None:
            print("No table on the page, exiting!")
            break

        table_data_single = []

        # headers are indicators of a quarter
        headers = table.find('thead').find_all('th')
        headers_data = [header.get_text() for header in headers]
        table_data_single.append(headers_data)
        
        #print(len(headers))
        #for header in headers:
        #    print(header.get_text())

        table_rows = table.find('tbody').find_all('tr')
        
        for table_row in table_rows:
            row_datas = table_row.find_all('td')
            row_data = [remove_spaces(row_dat.get_text(separator='\n')) for row_dat in row_datas]
            table_data_single.append(row_data)
            #for row in table_data_single:
            #    print(row)
        page_number += 1 

        # table transposition
        table_data_single_tr = [[row[i] for row in table_data_single] for i in range(len(table_data_single[0]))]

        table_data += table_data_single_tr[1:]
        #for row in table_data:
        #    print(row)

    # write result to a file
    csv_file = 'data.csv'
    match = re.search(r'/akcje/([^/]+)/wyniki-finansowe', wyniki_href)

    if match:
        csv_file = match.group(1)
        csv_file = csv_file + ".csv"

    data_dir = 'data_from_scrapper'
    file_path = os.path.join(data_dir, csv_file)

    with open(file_path, 'w', newline='') as file:
        writer = csv.writer(file)
        for row in table_data:
            writer.writerow(row)
    
```
